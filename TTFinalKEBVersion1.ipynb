Sometimes they get interpolated where they don't belong (e.g. lavinier) recreation --> récréation is interpolating the modern spelling, does that mean OCR is drawing from a database of modern french?

Tilde contractions either don't go through(grãd for grand, becomes grad) or are converted to other accents (expirēt for expirent, becomes expirét). Larger question: how to deal with these as part of a word count/word sort? Even larger question: are words that get contracted likely to be key words?
